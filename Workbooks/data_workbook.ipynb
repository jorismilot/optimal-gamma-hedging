```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import numpy as np

path = '/Users/joris/Documents/Master QF/Thesis/optima-gamma-hedging/Data/snapshot_data'
coins = ['btc'] # , 'eth'

for coin in coins: 
    df_raw = pd.read_json(os.path.join(path, f'{coin}_full_data.json'))
```

```python
def is_single_dict(cell):
    return isinstance(cell, dict)

# Apply the function to all cells
mask = df_raw.map(is_single_dict)

# Check if all non-null entries are single dicts
all_single_dicts = mask.where(df_raw.notna()).all().all()

print("All non-null cells are exactly one dictionary:", all_single_dicts)
```
